```python
import pandas as pd
df=pd.read_csv("fake.csv")
print(df.describe())
```

```python
#missing values
print(df.isnull().sum())
```

```python
#type distribution
print(df['type'].value_counts())
```

```python
#histogram

import seaborn as sns
import matplotlib.pyplot as plt

df.dropna().hist(figsize=(10,8))
plt.show()

sns.pairplot(df.dropna().drop_duplicates(),hue='type')
plt.show()
```

```python
#bar plots

sns.boxplot(data=df.dropna().drop_duplicates())
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv("fake.csv")
cols=['title','text','domain_rank','country','type']
df_new=df[cols].copy()
df_new['domain_rank'] = df_new['domain_rank'].fillna(df_new['domain_rank'].median())
df_new['country'] = df_new['country'].fillna("unknown")
df_new = df_new.dropna(subset=['title','text'])



sns.pairplot(df_new,hue='type')
plt.show()
```
